```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../raw_data/train.csv')
```

```python
plt.hist(df['Basic_Demos-Age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
plt.hist(df['Physical-BMI'], bins=20)
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('Distribution of BMI')
plt.show()
```

```python
sns.scatterplot(x='Physical-Height', y='Physical-Weight', data=df)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs. Weight')
plt.show()
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.lineplot(x='Basic_Demos-Enroll_Season', y='Physical-BMI', data=df)
plt.xlabel('Season')
plt.ylabel('BMI')
plt.title('BMI Over Seasons')
plt.show()
```
